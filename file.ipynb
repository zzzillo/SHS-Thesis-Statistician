## Likert Scale Outputs

```python
import pandas as pd

likert_labels = {1: "SD", 2: "D", 3: "N", 4: "A", 5: "SA"}

def get_likert_results(column_data):
    freq = column_data.value_counts().reindex(likert_labels.keys(), fill_value=0)
    percent = (freq / freq.sum()) * 100
    
    df = pd.DataFrame({
        "Scale": [likert_labels[i] for i in freq.index],
        "Code": freq.index,
        "Frequency": freq.values,
        "Percentage": percent.round(2).values
    })
    df.loc[len(df)] = ["Total", "", freq.sum(), 100]
    return df

def print_stats(column_data, label):
    print(f"--- {label} Statistics ---")
    print(f"Mean: {round(column_data.mean(), 2)}")
    print(f"Median: {column_data.median()}")
    print(f"Mode: {column_data.mode()[0]} \n")
```

## Significant Relationship of GWA and WA

```python
columns = ["WA", "GWA"]

for col in columns:
    data = file2_clean[col].dropna()
    stat, p = shapiro(data)
    print(f"Shapiro-Wilk test for {col}: W={stat:.4f}, p-value={p:.4e}")
    if p > 0.05:
        print(f"{col} appears to be normally distributed.\n")
    else:
        print(f"{col} is NOT normally distributed.\n")
```

```python
data = file2_clean[["WA", "GWA"]].dropna()

rho, p_value = spearmanr(data["WA"], data["GWA"])

print(f"Spearman Correlation Coefficient (ρ): {rho:.4f}")
print(f"P-value: {p_value:.4e}")

# Interpretation
if p_value <= 0.05:
    print("The correlation is statistically significant. Accept the Null Hypothesis.")
else:
    print("The correlation is NOT statistically significant. Reject the Null Hypothesis.")
```

```python
rho = 0.1503
p_val = 0.0406
n = len(file2_clean[["WA", "GWA"]].dropna())
r_squared = rho**2
stderr = np.sqrt((1 - rho**2) / (n - 2))

print(f"Sample Size (n): {n}")
print(f"Spearman's Rho: {rho}")
print(f"P-value: {p_val}")
print(f"Coefficient of Determination (R²): {round(r_squared, 4)}")
print(f"Standard Error: {round(stderr, 4)}")
```

```python
import pandas as pd

def find_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# Find outliers for WA
outliers_wa, low_wa, high_wa = find_outliers(file2_clean, "WA")
# Find outliers for GWA
outliers_gwa, low_gwa, high_gwa = find_outliers(file2_clean, "GWA")

print(f"--- WA Outlier Bounds: {low_wa} to {high_wa} ---")
print(f"Number of WA outliers: {len(outliers_wa)}")
print(outliers_wa[["WA", "GWA"]].head())

print(f"\n--- GWA Outlier Bounds: {low_gwa} to {high_gwa} ---")
print(f"Number of GWA outliers: {len(outliers_gwa)}")
print(outliers_gwa[["WA", "GWA"]].head())
```
